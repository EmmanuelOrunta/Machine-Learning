```python
# Load the NSW covid-19 dataset from the provided URL - This code is meant to load the data from the given URL.
url = "https://data.nsw.gov.au/data/dataset/0a52e6c1-bc0b-48af-8b45-d791a6d8e289/resource/5200e552-0afb-4bde-b20f-f8dd4feff3d7/download/c19_location_09.24.csv"
df = pd.read_csv(url, encoding='latin-1')

# Display the first few rows - This code is to display the first 5 rows of the imported data, this gives us a preview of how the data looks like.
print("First 5 rows of the dataset:")
display(df.head())

# Below is the output of the code above showing the first 5 rows of my loaded dataset. From the dataset, I can see about 42 columns
 (dates, location, address and more) and some columns possess null values
```

```python
# Displaying basic information about the dataset - The code below gives us some in depth information of the dataset such as how many non-null
values each columns has and the datatypes of the respective columns

print("\nDataset Info:")
print(df.info())

# Summary statistics - The part of the code was used to describe the data using statistics fucntions such as mean, count, unique count, min,
max, standard deviation and more.

print("\nSummary Statistics:")
display(df.describe(include='all'))
```

```python
# Checking for missing values - I used this to check for how many NULL values each of my columns had.
print("\nMissing values in each column:")
print(df.isnull().sum())
```

```python
print(df.columns.tolist())  # See all column names
df.head(3)
```

```python
# Grouping by location and sorting from highest to lowest
cases_by_location = df.groupby("Location").size().sort_values(ascending=False)

print("\nTop locations by number of exposure records:")
print(cases_by_location.head(30))
```

```python
cases_by_location.head(10).plot(
    kind='bar',
    x='Location',
    y='Record_Count',
    legend=False,
    color='skyblue',
    figsize=(10,6)
)
plt.title("Top 10 Locations by COVID-19 Case Records")
plt.ylabel("Number of Records")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Step 9: Visualization - cases over time (if date column exists)

df["Date_1"] = pd.to_datetime(df["Date_1"], errors='coerce')
daily_cases = df.groupby("Date_1").size()

plt.figure(figsize=(12,6))
daily_cases.plot()
plt.title("COVID-19 Cases Over Time")
plt.ylabel("Number of Cases")
plt.xlabel("Date")
plt.show()
```

```python
# Task 3: Explore the Scikit-learn Digits Dataset

# Import necessary libraries
from sklearn import datasets        # to load the digits dataset
import matplotlib.pyplot as plt     # to visualize digit images
import numpy as np                  # for numerical operations

# 3.1 Load the Digits dataset
digits = datasets.load_digits()

# Print information about the dataset
print("Dataset Keys:", digits.keys())  # what information the dataset contains
print("Shape of images:", digits.images.shape)  # shape of the image data (number of samples, 8x8 pixels)
print("Shape of data (flattened images):", digits.data.shape)  # flattened images (each image = 64 features)
print("Shape of target:", digits.target.shape)  # labels for each image

# 3.2 Show a non-zero handwritten digit image
# Find the index of the first digit that is not zero
non_zero_index = np.where(digits.target != 0)[0][0]

# Display that digit image
plt.imshow(digits.images[non_zero_index], cmap='gray')  # show as grayscale image
plt.title(f"Digit: {digits.target[non_zero_index]}")    # display the digit label as title
plt.axis('off')  # hide the axes for clarity
plt.show()

# 3.3 Exploration of the dataset
# Show unique labels (digits 0–9)
print("Unique target labels:", np.unique(digits.target))

# Count how many samples exist for each digit
print("Number of samples per digit:", np.bincount(digits.target))

# Show first 10 target labels to understand data
print("First 10 target labels:", digits.target[:10])

# Visualize multiple digit samples to get a better overview
fig, axes = plt.subplots(2, 5, figsize=(10, 5))  # create a grid (2 rows, 5 columns)
for i, ax in enumerate(axes.flatten()):
    ax.imshow(digits.images[i], cmap='gray')  # display each digit image
    ax.set_title(f"Digit: {digits.target[i]}") # show its label
    ax.axis('off')  # hide the axes
plt.show()
```
